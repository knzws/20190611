```python
from sklearn import datasets
import pandas as pd

from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.metrics import mean_squared_error
```

```python
boston = datasets.load_boston()

X_RM = boston.data[:,[5]]
y_PRICE = boston.target
```

```python
clf = linear_model.LinearRegression()

clf.fit(X_RM, y_PRICE)
```

```python
a = clf.coef_  #係数
b = clf.intercept_  #切片

print('coefficient a =',a)
print('intercept b =',b)
```

```python
plt.scatter(X_RM, y_PRICE, c='black', alpha=0.5)

lreg_y = a * X_RM + b
plt.plot(X_RM, lreg_y, 'r')

plt.ylabel('Price in $1,000s')
plt.xlabel('Number of rooms')
plt.title('Boston house-prices')
```

```python
y_pred = clf.predict(X_RM)
mse = mean_squared_error(y_PRICE, y_pred)
print('MSE = ', mse)

print('R^2 = ', clf.score(X_RM, y_PRICE))
```
